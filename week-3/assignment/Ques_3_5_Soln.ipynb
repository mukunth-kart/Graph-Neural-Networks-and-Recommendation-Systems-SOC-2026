Activation functions are mainly used to introduce nonlinearity into a neural network. Since the identity function y = x is already a linear function, a single linear layer is sufficient.

Therefore, the simplest and most efficient architecture is:

Input → Linear(1,1) → Output

A single-neuron MLP with one linear layer is sufficient to perfectly represent the identity function. No activation function is required because the target function is linear. The exact solution is obtained when the weight equals 1 and the bias equals 0.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Create 100 evenly spaced values between -10 and 10
# Shape initially: (100,) #reshaping into (100,1) tensor
x = torch.linspace(-10, 10, 100).reshape(-1, 1)
y = x.clone()

# Simplest network: one linear layer, no activation
model = nn.Linear(1, 1)

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

for epoch in range(1000):
    optimizer.zero_grad()
    y_pred = model(x)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()

print("Final weight:", model.weight.data)
print("Final bias:", model.bias.data)
print("Final loss:", loss.item())
```

A single linear neuron has the form:

ŷ = wx + b

This represents a straight line.

However, the function y = x² is a curved parabola. A straight line cannot perfectly represent a curved function for all values of x.

Therefore, a single linear layer without an activation function is not sufficient to learn y = x².

**Need for Activation Function**

An activation function is needed because it introduces non-linearity into the network.

Without an activation function, even multiple linear layers collapse into one linear layer.

For example:

Linear → Linear → Linear

is still equivalent to one linear transformation.

So, to learn a non-linear function like y = x², the network must contain at least one non-linear activation function.

**Network Architecture**

A suitable simple architecture is:

Input → Linear Layer → ReLU Activation → Linear Layer → Output

The hidden layer allows the model to build a non-linear approximation of the function y = x².

**Expected Weights and Biases**

for y = x², the exact final weights are not simple or unique.

Many different combinations of weights and biases can approximate the same parabolic function.

Therefore, instead of expecting one exact set of weights, we expect the trained model to learn parameters that make the loss very small and produce outputs close to x².



```python
#importing PyTorch library
import torch

#importing neural network tools from PyTorch
import torch.nn as nn

#importing optimization algorithms
import torch.optim as optim


#Creating 100 input values between -10 and 10
x = torch.linspace(-10, 10, 100)

#reshaping x into a column vector of shape (100, 1)
#this means 100 samples, 1 feature per sample
x = x.reshape(-1, 1)

#creating target output using y = x^2
y = x ** 2


#defining an MLP for learning y = x^2
model = nn.Sequential(

    #First linear layer
    #Takes 1 input feature and maps it to 10 hidden neurons
    nn.Linear(1, 10),

    #ReLU activation function
    #Introduces non-linearity into the network
    nn.ReLU(),

    #Second linear layer
    #Takes 10 hidden values and produces 1 output
    nn.Linear(10, 1)
)


#Mean Squared Error loss function
#Measures average squared difference between prediction and actual y
criterion = nn.MSELoss()


#Adam optimizer
#Updates the weights and biases during training
optimizer = optim.Adam(model.parameters(), lr=0.01)


#Training loop
for epoch in range(5000):

    #Clear old gradients
    optimizer.zero_grad()

    #Forward pass: compute prediction
    y_pred = model(x)

    #Compute loss between prediction and actual y
    loss = criterion(y_pred, y)

    #Backward pass: compute gradients
    loss.backward()

    #Update weights and biases
    optimizer.step()


#Print final loss
print("Final loss:", loss.item())


#Print final weights and biases
for name, parameter in model.named_parameters():
    print(name, parameter.data)
```

```python
#New inputs that were not used during training
test_x = torch.tensor([
    [-7.5],
    [-3.0],
    [0.0],
    [2.5],
    [8.0]
])

#Disable gradient computation since we are only testing
with torch.no_grad():

    #Get model predictions
    predicted_y = model(test_x)

#True values according to y = x^2
actual_y = test_x ** 2

print("\nTesting on Fresh Data")
print("-" * 50)

for x_val, pred, actual in zip(test_x, predicted_y, actual_y):
    print(
        f"x = {x_val.item():6.2f} | "
        f"Predicted = {pred.item():8.3f} | "
        f"Actual = {actual.item():8.3f}"
    )
```

**Results**

The function y = x² is non-linear, so a single linear layer cannot represent it perfectly.

In this experiment, an MLP with one hidden layer and a ReLU activation function was used.

The ReLU activation function was necessary because it introduced non-linearity into the network. Without activation, multiple linear layers would still behave like a single linear layer.

After training, the loss became small, showing that the model learned to approximate the function y = x².

The final weights are not expected to be simple values like weight = 1 and bias = 0. This is because many different combinations of weights and biases can produce a good approximation of the parabola.

Therefore, the important observation is not whether the weights match one exact expected value, but whether the final loss is small and the predictions are close to y = x².

When training a neural network for classification tasks, two common combinations are:

1.Mean Squared Error (MSE) loss with a tanh activation function.


2.Cross-Entropy loss with a Softmax activation function.

The objective is to determine which combination performs better for classification problems.

Classification problems require the network to assign probabilities to different classes. The Softmax activation function converts the output logits into a probability distribution, ensuring that all output probabilities are between 0 and 1 and sum to 1.

The Cross-Entropy loss function is specifically designed to measure the difference between the predicted probability distribution and the true class labels. It provides strong gradients when predictions are incorrect, helping the network learn more efficiently and converge faster.

On the other hand, MSE is primarily intended for regression problems. When used with tanh for classification:

The output is constrained between -1 and 1 rather than representing probabilities.

Gradients can become very small when the tanh activation saturates, leading to slower learning.

The loss function does not directly measure classification error or probability mismatch.

## Mathematical Intuition

For a classification problem with (C) classes, the Softmax activation function converts the network outputs into probabilities.

### Softmax Output

$$
p_i = \frac{e^{z_i}}{\sum_{j=1}^{C} e^{z_j}}
$$

where:

* (z_i) is the output score for class (i)
* (C) is the total number of classes
* (p_i) is the predicted probability for class (i)

The Softmax function ensures that:

$$
0 \leq p_i \leq 1
$$

and

$$
\sum_{i=1}^{C} p_i = 1
$$

Thus, the outputs can be interpreted as probabilities.

### Cross-Entropy Loss

The Cross-Entropy loss is defined as:

$$
L = -\sum_{i=1}^{C} y_i \log(p_i)
$$

where:

* (y_i) is the true label
* (p_i) is the predicted probability for class (i)

### Interpretation

Cross-Entropy directly measures how different the predicted probability distribution is from the true distribution.

If the network assigns a high probability to the correct class, the loss becomes very small.

If the network assigns a low probability to the correct class, the loss becomes large.

Therefore, Cross-Entropy loss encourages the network to assign high probability to the correct class and low probability to incorrect classes, leading to better classification performance.
